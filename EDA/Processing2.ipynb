```python
import pandas as pd
import numpy as np
from collections import Counter
import pickle
import os
```

```python
## Load pickle AI 리스트 불러오기
with open("artificial_user_list","rb") as fr:
    pik_bot_id = pickle.load(fr)
print(pik_bot_id)
```

```python
# 데이터 불러오기

df1 = pd.read_csv('piks_pik.csv')
df2 = pd.read_csv('piks_category.csv')
df3 = pd.read_csv('linkhub_link.csv')
```

```python
# pikurate bot del
def pik_bot_del (data, col) :
    with open("artificial_user_list","rb") as fr:
        pik_bot_id = pickle.load(fr)
    new_df = data[~data[col].isin(pik_bot_id)]
    print(f"피큐봇 제거 전 데이터 개수 : {data.shape[0]}개")
    print(f"피큐봇 제거 후 데이터 개수 : {new_df.shape[0]}개")
    return new_df
```

```python
Picks = pik_bot_del (df1, "user_id")
```

```python
Cates = pik_bot_del (df2, "user_id")
```

```python
Links = pik_bot_del (df3, "user_id")
```

```python
df1
```

```python
link_col = ["Link_id", "category_id"]
Links.rename(columns = {'id' : 'Link_id'}, inplace=True)
Cates.rename(columns = {'id' : 'Category_id'}, inplace=True)
new_Links = Links.loc[:, link_col]
new_Links = new_Links.groupby("category_id").count()
new_Links
```

```python
Cates
```

```python
merged_df = pd.merge(Cates.loc[:, ["pik_id", "Category_id"]], new_Links, how='left', left_on='Category_id', right_index = True)

# 결과 확인
a = merged_df.groupby('pik_id').count()["Category_id"]
b = merged_df.groupby('pik_id').sum()["Link_id"]
pik_count = pd.merge(a, b, right_index=True, left_index=True)
pik_count
```

```python
Picks2 = Picks.loc[(Picks['is_default'] == False)&(Picks['is_deleted'] == False)]
Picks2
```

```python
new_piks2 = pd.merge(Picks2.iloc[:, [0, 1, 4, 8, 9]], pik_count, how="left", left_on="id", right_index=True)
new_piks2.to_csv("new_piks2.csv")
```
